```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the CSV dataset

```python
file_path = r"C:\Users\jeyka\INTERNSHIP 2025\SKILLCRAFT TECH JULY'25\SCT_DS_TASK4\accident_dataset\US_Accidents_March23.csv"
df = pd.read_csv(file_path)
```

# Step 1: Convert date and extract features

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df['Hour'] = df['Start_Time'].dt.hour
df['DayOfWeek'] = df['Start_Time'].dt.day_name()
```

# Step 2: Accidents by Hour

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='Hour', hue='Hour', palette='viridis', legend=False)
plt.title(" Accidents by Hour of Day")
plt.ylabel("Accident Count")
plt.xlabel("Hour")
plt.tight_layout()
plt.show()
```

# Step 3: Accidents by Day of Week

```python
plt.figure(figsize=(10, 4))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(data=df, x='DayOfWeek', hue='DayOfWeek', order=order, palette='magma', legend=False)
plt.title("Bookings by Day of Week")  
plt.tight_layout()
plt.show()
```

# Step 4: Accidents by Weather Condition

```python
plt.figure(figsize=(12, 5))
df['Weather_Condition'].value_counts().nlargest(10).plot(kind='bar', color='teal')
plt.title(" Top 10 Weather Conditions During Accidents")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Step 5: Severity Distribution

```python
import pandas as pd
severity_counts = df['Severity'].value_counts().sort_index()
import numpy as np

import matplotlib.pyplot as plt
severity_counts.plot(kind='bar', color=plt.cm.coolwarm(np.linspace(0, 1, len(severity_counts))))
plt.title("Count of Incidents by Severity")
plt.xlabel("Severity Level")
plt.ylabel("Number of Incidents")
plt.tight_layout()
plt.show()
```
