```python
import matplotlib.pyplot as plt

# Data based on the image for India's population distribution in 2022
age_groups = ["0–20 Years", "21–64 Years", "65+ Years"]
population_millions = [512, 807, 98]

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(age_groups, population_millions, color=['gold', 'dodgerblue', 'hotpink'])

# Add population values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 5, f'{yval} Mn', ha='center', va='bottom', fontsize=12)

# Chart title and labels
plt.title("India's Population Distribution by Age Group (2022)", fontsize=14)
plt.ylabel("Population (Millions)")
plt.xlabel("Age Group")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the plot
plt.show()
```
